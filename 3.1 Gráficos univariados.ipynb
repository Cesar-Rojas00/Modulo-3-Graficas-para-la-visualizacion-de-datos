```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
netflix_data = pd.read_csv('netflix_titles.csv')
```

```python
netflix_data.head()
```

```python
# Histograma para la variable 'release_year'
plt.hist(netflix_data['release_year'], bins = 30, color = 'red', edgecolor = 'black')
plt.title('Distribución de los años de lanzamiento')
plt.xlabel('Año de lanzamiento')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Crear gráfico de barras
sns.countplot(data = netflix_data, x = 'type')
plt.title('Cantidad de peliculas y series')
plt.xlabel('Tipo')
plt.ylabel('Cantidad')
plt.show()
```

```python
# Crear gráfico circular
plt.pie(netflix_data['type'].value_counts(), labels = netflix_data['type'].value_counts().index, autopct = '%1.1f%%')
plt.title('Porcentaje de peliculas y series')
plt.show()
```

```python
# Crear gráfico de densidad
sns.kdeplot(data = netflix_data['release_year'], fill = True, color = 'red')
plt.title('Densidad de los años de lanzamiento')
plt.xlabel('Año de Lanzamiento')
plt.ylabel('Densidad')
plt.show()
```
