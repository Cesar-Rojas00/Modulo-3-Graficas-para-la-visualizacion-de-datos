```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
netflix_data = pd.read_csv('netflix_titles.csv')
netflix_data.head()
```

```python
# Procesamiento de los datos. Limpieza de los datos
# Eliminar los valores nulos
netflix_data.dropna(subset = ['date_added'], inplace = True)
```

```python
netflix_data.head()
```

```python
# Es buena practica enteder la estrucutura de nuesto dataset
netflix_data.info()
```

```python
# Reemplazar los valores nulos con Unknowkn
netflix_data = netflix_data.fillna('Unknown')
```

```python
netflix_data.info()
```

```python
# Gráfico de barras para comparar por año las peliculas o series
plt.figure(figsize = (12, 8))
sns.countplot(x = 'release_year', data = netflix_data, hue = 'type', palette = 'Set2')
plt.title('Cantidad de peliculas y series por año')
plt.xlabel('Año')
plt.ylabel('Cantidad')
plt.xticks(rotation = 90)
plt.show()
```

```python
# Convertir una variable a formato de fecha
netflix_data['date_added'] = pd.to_datetime(netflix_data['date_added'].str.strip(), format='%B %d, %Y', errors='coerce')
```

```python
netflix_data.dtypes
```

```python
netflix_data.sample(5)
```

```python
# Visualización del crecimiento de generos en Netflix a lo largo del tiempo
# Nueva columna con año
netflix_data['year_added'] = pd.to_datetime(netflix_data['date_added']).dt.year
```

```python
# Separar los generos en filas individuales
netflix_data['listed_in'] = netflix_data['listed_in'].str.split(',')
```

```python
# Crear un dataframe auxiliar con los generos individuales
genre_data = netflix_data.explode('listed_in')
```

```python
# Obtener conteo de genero por año
genre_counts = genre_data.groupby('year_added')['listed_in'].value_counts().unstack().fillna(0)
```

```python
# Graficar el conteo a lo largo del tiempo
genre_counts.plot(kind = 'line', figsize = (12, 8))
plt.title('Crecimiento de géneros en Netflix a lo largo del tiempo')
plt.xlabel('Año')
plt.ylabel('Conteo')
plt.legend(loc = 'center left', bbox_to_anchor = (1, 0.5))
plt.show()
```

```python
# Analisis de generos seleccionados
selected_genres = ['Action & Adventure', 'Comedies', 'Dramas', 'Documentaries', 'Romantic Movies', 
                  'Thrillers', 'Children & Family Movies', 'Sci-Fi & Fantasy', 'Horror Movies', 'International Movies']
```

```python
genre_data_filtered = genre_data[genre_data['listed_in'].isin(selected_genres)]
```

```python
genre_counts = genre_data_filtered.groupby('year_added')['listed_in'].value_counts().unstack(fill_value = 0)
genre_counts.plot(kind = 'line', figsize = (12, 8))
plt.title('Crecimiento de géneros en Netflix a lo largo del tiempo')
plt.xlabel('Año')
plt.ylabel('Conteo')
plt.legend(loc = 'center left', bbox_to_anchor = (1, 0.5))
plt.show()
```
